```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def log_map(x0,r):
    x1 = 4*r*x0*(1-x0)
    return x1
```

```python
x_ini = 1/np.pi

def log_map_fin(x_ini,r,N_iter):
    x = np.zeros((N_iter))
    x[0] = x_ini
    for i in range(N_iter-1):
        x[i+1] = log_map(x[i],r)
        
    return x[-32:]
```

```python
N_iter = 300
N_r = 2000
r = np.linspace(0.7,1,N_r)

for i in range(N_r):
    a = log_map_fin(x_ini,r[i],N_iter)
    for j in range(len(a)):
        plt.scatter(r[i],a[j],color='black',s=0.3)
plt.savefig('log_map.png',dpi=1000,bbox_inches='tight')        
plt.show()    
```
